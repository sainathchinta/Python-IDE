```python
from tkinter import *
from tkinter import messagebox
from tkinter.filedialog import asksaveasfilename ,askopenfilename
import subprocess 
import os

root=Tk()
root.title("Python IDE")
root.geometry("1280x720+150+80")
root.configure(bg="#323846")
root.resizable(False,False)

filepath=''

def setfilepath(path):
    global filepath
    filepath=path
    

def open_file():
    path=askopenfilename(filetypes=[('Python Files','*.py')])
    with open(path,'r') as file:
        code=file.read()
        inputcode.delete('1.0',END)
        inputcode.insert('1.0',code)
        setfilepath(path)
    
def save_file():
    if filepath=='':
        path= asksaveasfilename(filetypes=[('Python Files','*.py')])
    else:
        path=filepath
    
    with open(path,'w') as file:
        code=inputcode.get('1.0',END)
        file.write(code)
        setfilepath(path)
    
    
        
    
def run_file():
    if filepath=='':
        messagebox.showerror("first save code")
        return
    code=inputcode.get('1.0',END)
    exec(code)
    outputcode.insert('1.0',code)
  
   
    

#icon
icon=PhotoImage(file="logo.png")
root.iconphoto(False,icon)

#input
inputcode=Text(root,font="cosolas 18")
inputcode.place(x=180,y=0,width=680,height=720)

#output
outputcode=Text(root,font="cosolas 15",fg="lightgreen",bg="#323846")
outputcode.place(x=860,y=0,width=420,height=720)

#buttonsimages
openi=PhotoImage(file="open.png")
runi=PhotoImage(file="run.png")
savei=PhotoImage(file="save.png")



openbtn=Button(root,image=openi,bg="#323846",bd=0,command=open_file)
openbtn.place(x=30,y=30)

savebtn=Button(root,image=savei,bg="#323846",bd=0,command=save_file)
savebtn.place(x=30,y=180)

runbtn=Button(root,image=runi,bg="#323846",bd=0,command=run_file)
runbtn.place(x=30,y=330)

root.mainloop()


```
